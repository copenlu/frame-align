```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Topic Samples

```python
sample_df = pd.read_csv("/projects/copenlu/data/arnav/frame-align/raw/2023-24/topic_samples.csv")
sample_anno_df = pd.read_json("/home/ksg502/projects/frame-align/data/annotated/text/topic_samples_Mistral.jsonl", lines=True)
```

```python
merged_df = sample_df.merge(sample_anno_df, on="id")
```

```python
merged_df['topic_label'].nunique()
```

```python
plt.subplots(figsize=(10, 70))
plt.tick_params(axis='x', which='major', pad=25)
for i, (group_name, group) in enumerate(merged_df.groupby('topic_label')):
    plt.subplot(15, 5, i+1)
    group['frame_name'].value_counts().plot(kind='bar', xlabel=None)
    plt.title(group_name)
    plt.tight_layout()
plt.savefig("../data/figures/topic_sample_frames.png")
```
